## gradiente sin perdida de datos

```python
def calcular_gradiente(datos, marco):
    pesos = np.array([marco - i for i in range(marco)])  
    pesos = pesos / np.sum(pesos)                        
    niveles_restantes = datos.shape[0] - 1 - 2 * (marco - 1)
    gradiente_datos = np.zeros((niveles_restantes, datos.shape[1]))

    for i in range(marco, datos.shape[0] - marco):
        superior = np.sum([pesos[j] * datos[i + j, :] for j in range(marco)], axis=0)
        inferior = np.sum([pesos[j] * datos[i - j - 1, :] for j in range(marco)], axis=0)
        gradiente_datos[i - marco, :] = superior - inferior

    filas_superior = marco
    filas_inferior = marco - 1

    gradiente_datos_completo = np.pad(gradiente_datos, ((filas_superior, filas_inferior), (0, 0)), mode='constant', constant_values=0)

    return gradiente_datos_completo
```

# Codigo para detectar isoterma

```python
def detectar_isoterma_cero(gradiente, altura_inicial):
    resultados = []
    altura_actual = altura_inicial

    for t in range(gradiente.shape[1]):                      # Recorrer cada columna de tiempo
        idx_altura = int((np.abs(heights_ajustado - altura_actual)).argmin()) #buscamos la coincidencia de la altura con el indice
        gradiente_minimo = gradiente[idx_altura, t]
        altura_minima = heights_ajustado[idx_altura].values  # Guardamos la altura correspondiente a este indice

        # Definimos los limites de busqueda (ventana) que en este caso seria de +-1 para comparar
        idx_superior = max(idx_altura - 1, 0)
        idx_inferior = min(idx_altura + 1, len(heights_ajustado) - 1)

        for i in range(idx_superior, idx_inferior + 1):     # Iteraciones dentro de la ventana
            if gradiente[i, t] < gradiente_minimo:
                gradiente_minimo = gradiente[i, t]
                altura_minima = heights_ajustado[i].values  # Actualizamos la altura del mínimo
        # descomentar para pruebas de valores
        #if altura_minima == altura_actual:
        #    print(f"Iter {t}: Sin cambio en la altura, se mantiene en {altura_actual}")
        #else:
        #    print(f"Iter {t}: Altura mínima ajustada a {altura_minima}, Gradiente mínimo {gradiente_minimo}")


        
        altura_actual = altura_minima                      # Actualizamos la altura para la iteracion siguiente
        resultados.append({
            "iter": t,
            "altura_minima": altura_actual,
            "gradiente_minimo": gradiente_minimo
        })

    return resultados

# Ejemplo de uso
altura_inicial = 2000
resultados_isoterma = detectar_isoterma_cero(gradiente_Ze, altura_inicial)
#print(resultados_isoterma)  # Comentado para evitar salida directa

```

## Filtro Kalman2

```python
# Este modelo consiste en usar mediciones ya obtenidas sobre la isoterma, usando esto para ser recalculados con filtro kalman
# para esto, usamos la covarianza de estos datos para el filtro kalman, multiplicado por un factor para poder controlar que tanto margen
# puede llegar a tener el modelo, usando factor_ponderador

def filtro_kalman2(alturas, delta_t, Q_ruido_proceso_reflectividad, Q_ruido_proceso_gradiente, R_ruido_medicion, factor_ponderador):
    from filterpy.kalman import KalmanFilter
    import numpy as np

    # Inicializar filtro Kalman
    f = KalmanFilter(dim_x=2, dim_z=1)
    f.x = np.array([alturas[0], 0])  # Estado inicial: [posición, velocidad (0 si no se usa)]
    f.F = np.array([[1, delta_t], [0, 1]])  # Modelo de transición
    f.H = np.array([[1, 0]])  # Matriz de observación (solo posición)
    f.P *= 0.1  # Matriz de covarianza inicial
    f.R = R_ruido_medicion  # Ruido de medición
    f.Q = np.array([[Q_ruido_proceso_reflectividad, 0], [0, Q_ruido_proceso_gradiente]])  # Ruido del proceso

    # Calcular varianza global de los datos originales
    varianza_global = np.var(alturas)
    desviacion_global = np.sqrt(varianza_global)

    # Lista para almacenar las alturas filtradas
    alturas_filtradas = []

    # Inicializar valores confiables para control de varianza
    valores_confiables = [alturas[0]]

    for z in alturas[1:]:
        # Predecir el siguiente estado
        f.predict()

        # Calcular media de los últimos valores confiables
        media = np.mean(valores_confiables[-10:])  # Últimos 10 valores confiables

        # Determinar límites basados en varianza global y factor_ponderador
        limite_varianza_inferior = media - factor_ponderador * desviacion_global
        limite_varianza_superior = media + factor_ponderador * desviacion_global

        # Aplicar criterios de filtrado
        if limite_varianza_inferior <= z <= limite_varianza_superior:
            f.update(z)  # Actualizar con el valor confiable
            valores_confiables.append(z)  # Agregar a la lista de confiables

        # Almacenar la estimación actual
        alturas_filtradas.append(f.x[0].copy())

    return np.array(alturas_filtradas)


```

```python
# Datos originales ajustados
altura_inicial_desfase = 500 + (heights[1].values - heights[0].values) / 2
AlturasIsoterma_ajustadas = AlturasIsoterma + altura_inicial_desfase

# Parámetros del filtro Kalman
delta_t = 20  # Intervalo de tiempo
Q_ruido_proceso_reflectividad = 1e-5   # Ruido del proceso para reflectividad
Q_ruido_proceso_gradiente = 1e-5       # Ruido del proceso para gradiente
R_ruido_medicion = 1                   # Ruido de medición
factor_ponderador = 0.65               # Ponderador para límites dinámicos(aca se modifica para ver la tolerancia del modelo)

# Aplicar el filtro Kalman
alturas_filtradas = filtro_kalman2(AlturasIsoterma_ajustadas, 
                                   delta_t, 
                                   Q_ruido_proceso_reflectividad, 
                                   Q_ruido_proceso_gradiente, 
                                   R_ruido_medicion, 
                                   factor_ponderador)
alturas_filtradas
```

# Aplicar Detector de isoterma sobre gradiente con restriccion de altura

```python
def detectar_isoterma_cero_limitado(gradiente, altura_inicial, salto_maximo):
    resultados = []
    altura_actual = altura_inicial

    for t in range(gradiente.shape[1]):                                       # Recorrer cada columna de tiempo 
        idx_altura = int((np.abs(heights_ajustado - altura_actual)).argmin()) # Ajuste que guarda
        idx_superior = max(idx_altura - 1, 0)
        idx_inferior = min(idx_altura + 1, len(heights_ajustado) - 1)
        idx_min = idx_superior + np.argmin(gradiente[idx_superior:idx_inferior + 1, t])
        altura_minima = heights_ajustado[idx_min].values

       
        if abs(altura_minima - altura_actual) > salto_maximo:                 # Aplicamos la restriccion de salto maximo
            altura_minima = altura_actual
        altura_actual = altura_minima
        resultados.append(idx_min)

    return resultados
```

```python
altura_inicial = 2000
salto_maximo = 200       # Máximo cambio permitido en altura (msnm)
indices_isoterma_limitado = detectar_isoterma_cero_limitado(gradiente_Ze, altura_inicial, salto_maximo)
gradiente_modificado_limitado = np.copy(gradiente_Ze)

for t, idx_altura in enumerate(indices_isoterma_limitado):
    gradiente_modificado_limitado[idx_altura, t] = 30
plot_mrr2(xlim, new_time, heights_ajustado, gradiente_modificado_limitado, hora_local=False)
```
